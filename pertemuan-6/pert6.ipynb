```python
from collections import Counter
```

## use our data

```python
ids = data['Responder_id']
lang_response = data['LanguagesWorkedWith']

lang_counter = Counter()
```

```python
for response in lang_response:
    lang_counter.update(response.split(';'))
```

```python
print(lang_counter)
```

```python
print(lang_counter.most_common(15))
```

```python
language = []
popularity = []

for item in lang_counter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
```

```python
print(language)
print(popularity)
```

## add to bar chart

```python
plt.bar(language, popularity)
plt.title('Most Popular Language')
plt.xlabel('Language')
plt.ylabel('Users')
plt.show()
```

```python
plt.bar(language, popularity)
plt.title('Most Popular Language')
plt.xlabel('Language')
plt.ylabel('Users')
plt.xticks(rotation=70)
plt.show()
```

```python
plt.barh(language, popularity)
plt.title('Most Popular Language')
plt.xlabel('Language')
# plt.ylabel('Users')
plt.show()
```

```python
language.reverse()
popularity.reverse()
plt.barh(language, popularity)
plt.title('Most Popular Language')
plt.xlabel('Language')
# plt.ylabel('Users')
plt.show()
```
