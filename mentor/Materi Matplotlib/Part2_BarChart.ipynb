# Use our data

```python
ids = data['Responder_id']
lang_response = data['LanguagesWorkedWith']

language_counter = Counter()
```

```python
for response in lang_response:
    language_counter.update(response.split(';'))

print(language_counter.most_common(15))
```

```python
language = []
popularity = []

for item in language_counter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
```

```python
print(language)
print(popularity)
```

# Add to bar chart

```python
plt.bar(language, popularity)

plt.title("Most Popular Language")
plt.xlabel("Language")
plt.ylabel("Users")

plt.tight_layout()
plt.show()
```

When we have lots of category, it is better to use horizontal bar

```python
plt.barh(language, popularity)

plt.title("Most Popular Language")
plt.xlabel("Language")
#plt.ylabel("Users")

plt.tight_layout()
plt.show()
```

Reverse the order

```python
popularity.reverse()
language.reverse()

plt.barh(language, popularity)

plt.title("Most Popular Language")
plt.xlabel("Language")
#plt.ylabel("Users")

plt.tight_layout()
plt.show()
```
