```python
from matplotlib import pyplot as plt
```

```python
slices = [40, 60]
```

```python
plt.pie(slices)

plt.title("My First Pie Chart")
plt.tight_layout()
plt.show()
```

```python
labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels)

plt.title("My First Pie Chart")
plt.tight_layout()
plt.show()
```

The total can be beyond or lower than 100, the mlp will automatically calculate it

```python
slices=[90, 20]

labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels)

plt.title("My First Pie Chart")
plt.tight_layout()
plt.show()
```

We can give outlines

```python
slices=[90, 20]

labels = ['Sixty', 'Forty']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'k'})

plt.title("My First Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices=[90, 20, 30, 50]

labels = ['Sixty', 'Forty', 'add', 'new']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'k'})

plt.title("My First Pie Chart")
plt.tight_layout()
plt.show()
```

Use data of Number of People surveyed by StackOverflow

```python
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
```

```python
plt.pie(slices, labels=labels)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

It is not good to use pie chart for data that has more than 5 items

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(slices, labels=labels)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

We can emphasize an item

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]

plt.pie(slices, labels=labels, explode=explode)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

Add shadow

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]

plt.pie(slices, labels=labels, explode=explode, shadow=True)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

Start at certain point

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90)

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```

add percentage

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct = '%1.1f%%')

plt.title("My Pie Chart")
plt.tight_layout()
plt.show()
```
