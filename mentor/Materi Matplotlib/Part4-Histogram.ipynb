```python
import pandas as pd
from matplotlib import pyplot as plt
```

Let's assume that we conducted a survey. We have the data of the respondents' age and the amout of responden we have for each age

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
```

We certainly can visualize it by using bar chart. But, how if we have a hunded data? There would be too many bar. That's why we need histogram.
Histogram allows us to create bins for our data

```python
plt.hist(ages, bins=5)

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

Let's add a edgecolor

```python
plt.hist(ages, bins=5, edgecolor='k')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

We can pass our own list of value (bins). It allows us to have control

```python
# Plot the age in range of 10 years

ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 20, 30, 40, 50, 60]


plt.hist(ages, bins=bins, edgecolor='k')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

With this format, we can exclude some data

```python
# Plot the age in range of 10 years

ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [20, 30, 40, 50, 60]


plt.hist(ages, bins=bins, edgecolor='k')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

```python
ids = data['Responder_id']
ages = data['Age']

bins = [10, 20, 30, 40, 50, 60, 80, 90, 100]

plt.hist(ages, bins=bins, edgecolor='k', log=True)

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

Add certain line. In this case we use median_age=29

Use vertical line

```python
median_age = 29
color = 'red'
logg = 100

plt.hist(ages, bins=bins, edgecolor='k', log=True)

plt.axvline(median_age, color=color, label='Median Age')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

Use horizontal line

```python
median_age = 29
color = 'red'
logg = 100

plt.hist(ages, bins=bins, edgecolor='k', log=True)

plt.axvline(median_age, color=color, label='Median Age')
plt.axhline(logg, color='cyan', label='100 Number')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```
